```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
single_transformer_losses=[]
ltm_transformer_losses=[]

with open('/home/sauvageduck24/Escritorio/LTM/losses/single_transformer.txt') as f:
    [single_transformer_losses.append(float(line)) for line in f.readlines()]

with open('/home/sauvageduck24/Escritorio/LTM/losses/transformer_with_ltm.txt') as f:
    [ltm_transformer_losses.append(float(line)) for line in f.readlines()]
```

```python
df_single=pd.DataFrame(single_transformer_losses)
df_ltm=pd.DataFrame(ltm_transformer_losses)
```

```python
plt.figure(figsize=(16,8))

df_single[0].plot(label='Simple Transformer',color='red')
df_ltm[0].plot(label='Transformer with LTM',color='green')

plt.title('Comparation between LTM and Simple Transformer losses')
plt.xlabel('Iterations')
plt.legend()
```
